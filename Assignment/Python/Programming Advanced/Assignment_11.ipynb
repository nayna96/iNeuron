### 1. Create a function that takes a list and returns a new list containing only prime numbers.

**Examples**

filter_primes([7, 9, 3, 9, 10, 11, 27]) ➞ [7, 3, 11]<br>
filter_primes([10007, 1009, 1007, 27, 147, 77, 1001, 70]) ➞ [10007, 1009]<br>
filter_primes([1009, 10, 10, 10, 3, 33, 9, 4, 1, 61, 63, 69, 1087, 1091, 1093, 1097]) ➞ [1009, 3, 61, 1087, 1091, 1093, 1097]

**Ans:**

```python
def isPrime(n) :
    if (n <= 1) :
        return False
    
    if (n <= 3) :
        return True

    if (n % 2 == 0 or n % 3 == 0) :
        return False

    i = 5
    while(i * i <= n) :
        if (n % i == 0 or n % (i + 2) == 0) :
            return False
        i = i + 6

    return True

def filter_primes(lst):
    result = filter(lambda x: isPrime(x), lst)
    return list(result)

print(filter_primes([7, 9, 3, 9, 10, 11, 27]))
print(filter_primes([10007, 1009, 1007, 27, 147, 77, 1001, 70]))
print(filter_primes([1009, 10, 10, 10, 3, 33, 9, 4, 1, 61, 63, 69, 1087, 1091, 1093, 1097]))
```

### 2. Once a water balloon pops, is soaks the area around it. The ground gets drier the further away you travel from the balloon.

The effect of a water balloon popping can be modeled using a list. Create a function that takes a list which takes the pre-pop state and returns the state after the balloon is popped. The pre-pop state will contain at most a single balloon, whose size is represented by the only non-zero element.

**Examples**

pop([0, 0, 0, 0, 4, 0, 0, 0, 0]) ➞ [0, 1, 2, 3, 4, 3, 2, 1, 0]<br>
pop([0, 0, 0, 3, 0, 0, 0]) ➞ [0, 1, 2, 3, 2, 1, 0]<br>
pop([0, 0, 2, 0, 0]) ➞ [0, 1, 2, 1, 0]<br>
pop([0]) ➞ [0]

**Ans:**

```python
import math

def pop(lst):
    l = []

    n = math.floor(len(lst)/2)

    for i in range(n + 1):
        l.append(i)
        
    for i in range(n - 1 , -1, -1):
        l.append(i)

        
    return l
    
print(pop([0, 0, 0, 0, 4, 0, 0, 0, 0]))
print(pop([0, 0, 0, 3, 0, 0, 0]))
print(pop([0, 0, 2, 0, 0]))
print(pop([0]))
```

### 3. "Loves me, loves me not" is a traditional game in which a person plucks off all the petals of a flower one by one, saying the phrase "Loves me" and "Loves me not" when determining whether the one that they love, loves them back.

Given a number of petals, return a string which repeats the phrases "Loves me" and "Loves me not" for every alternating petal, and return the last phrase in all caps. Remember to put a comma and space between phrases.

**Examples**

loves_me(3) ➞ "Loves me, Loves me not, LOVES ME"<br>
loves_me(6) ➞ "Loves me, Loves me not, Loves me, Loves me not, Loves me, LOVES ME NOT"<br>
loves_me(1) ➞ "LOVES ME"

**Ans:**

```python
def loves_me(num):
    l = ["Loves me", "Loves me not"]
    lst = []
    
    for i in range(num):
        if i >= len(l):            
            lst.append(l[i % len(l)])
        else:
            lst.append(l[i])
    
    lst[-1] = lst[-1].upper()
    
    return ",".join(lst)

print(loves_me(3))
print(loves_me(6))
print(loves_me(1))
```

### 4. Write a function that sorts each string in a list by the letter in alphabetic ascending order (a-z).

**Examples**

sort_by_letter(["932c", "832u32", "2344b"])<br>
➞ ["2344b", "932c", "832u32"]

sort_by_letter(["99a", "78b", "c2345", "11d"])<br>
➞ ["99a", "78b", "c2345", "11d"]

sort_by_letter(["572z", "5y5", "304q2"])<br>
➞ ["304q2", "5y5", "572z"]

sort_by_letter([])<br>
➞ []

**Ans:**

```python
import re

def sort_by_letter(lst):
    import collections
    
    l = []
    dict_ = {}
    for el in lst:
        temp = re.findall(r'[a-zA-Z]+', el)
        dict_[temp[0]] = el

    od = collections.OrderedDict(sorted(dict_.items()))
    
    for k,v in od.items():
        l.append(v)
    return l
        
print(sort_by_letter(["932c", "832u32", "2344b"]))
print(sort_by_letter(["99a", "78b", "c2345", "11d"]))
print(sort_by_letter(["572z", "5y5", "304q2"]))
print(sort_by_letter([]))
```

### 5. There are three cups on a table, at positions A, B, and C. At the start, there is a ball hidden under the cup at position B.

However, I perform several swaps on the cups, which is notated as two letters. For example, if I swap the cups at positions A and B, I could notate this as AB or BA.

Create a function that returns the letter position that the ball is at, once I finish swapping the cups. The swaps will be given to you as a list.

**Examples**

cup_swapping(["AB", "CA", "AB"]) ➞ "C"

\# Ball begins at position B.<br>
\# Cups A and B swap, so the ball is at position A.<br>
\# Cups C and A swap, so the ball is at position C.<br>
\# Cups A and B swap, but the ball is at position C, so it doesn't move.

**Ans:**

```python
def cup_swapping(lst):
    cups = {'A': 0, 'B': 1, 'C': 0}
    
    for x, y in lst:
        cups[x], cups[y] = cups[y], cups[x]

    return {v:k for k, v in cups.items()}[1]

print(cup_swapping(["AB", "CA", "AB"]))
```
