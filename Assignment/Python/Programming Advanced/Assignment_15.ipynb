### 1. Write a function that returns True if a given name can generate an array of words.

**Examples**

anagram("Justin Bieber", ["injures", "ebb", "it"]) ➞ True<br>
anagram("Natalie Portman", ["ornamental", "pita"]) ➞ True<br>

anagram("Chris Pratt", ["chirps", "rat"]) ➞ False<br>
\# Not all letters are used

anagram("Jeff Goldblum", ["jog", "meld", "bluffs"]) ➞ False<br>
\# "s" does not exist in the original name

**Ans:**

```python
from collections import Counter

def anagram(name, arr):   
    x = list("".join(arr))
    y = list(name.replace(" ", "").lower())    
    return Counter(x) == Counter(y)

print(anagram("Justin Bieber", ["injures", "ebb", "it"]))
print(anagram("Natalie Portman", ["ornamental", "pita"]))
print(anagram("Chris Pratt", ["chirps", "rat"]))
print(anagram("Jeff Goldblum", ["jog", "meld", "bluffs"]))
```

### 3. Create a function that, given a phrase and a number of letters guessed, returns a string with hyphens - for every letter of the phrase not guessed, and each letter guessed in place.

**Examples**

hangman("helicopter", ["o", "e", "s"]) ➞ "-e---o--e-"<br>
hangman("tree", ["r", "t", "e"]) ➞ "tree"<br>
hangman("Python rules", ["a", "n", "p", "r", "z"]) ➞ "P----n r----"<br>
hangman("He"s a very naughty boy!", ["e", "a", "y"]) ➞ "-e"- a -e-y -a----y –y!"

**Ans:**

```python
def hangman(phrase, letters):
    l = list(phrase)
    lst = ["-" if i.isalnum() else i for i in phrase]
    for letter in letters:
        indices = [i for i, x in enumerate(l) if x.lower() == letter]        
        for index in indices:
            lst[index] = l[index]
    return "".join(lst)

print(hangman("helicopter", ["o", "e", "s"]))
print(hangman("tree", ["r", "t", "e"]))
print(hangman("Python rules", ["a", "n", "p", "r", "z"]))
print(hangman("He\"s a very naughty boy!", ["e", "a", "y"]))
```

### 4. The Collatz sequence is as follows:

- Start with some given integer n.
- If it is even, the next number will be n divided by 2.
- If it is odd, multiply it by 3 and add 1 to make the next number.
- The sequence stops when it reaches 1.

According to the Collatz conjecture, it will always reach 1. If that's true, you can construct a finite sequence following the aforementioned method for any given integer.

Write a function that takes in an integer n and returns the highest integer in the corresponding Collatz sequence.

**Examples**

max_collatz(10) ➞ 16<br>
\# Collatz sequence: 10, 5, 16, 8, 4, 2, 1

max_collatz(32) ➞ 32<br>
\# Collatz sequence: 32, 16, 8, 4, 2, 1

max_collatz(85) ➞ 256<br>
\# Collatz sequence: 85, 256, 128, 64, 32, 16, 8, 4, 2, 1

**Ans:**

```python
def max_collatz(n):
    l = []
    l.append(n)
    while l[-1] != 1:
        if l[-1] % 2 == 0:
            l.append(int(l[-1] / 2))
        else:
            l.append((l[-1] * 3) + 1)
    
    return max(l)

print(max_collatz(10))
print(max_collatz(32))
print(max_collatz(85))
```

### 5. Write a function that sorts a list of integers by their digit length in descending order, then settles ties by sorting numbers with the same digit length in ascending order.

**Examples**

digit_sort([77, 23, 5, 7, 101])<br>
➞ [101, 23, 77, 5, 7]

digit_sort([1, 5, 9, 2, 789, 563, 444])<br>
➞ [444, 563, 789, 1, 2, 5, 9]

digit_sort([53219, 3772, 564, 32, 1])<br>
➞ [53219, 3772, 564, 32, 1]

**Ans:**

```python
import collections

def digit_sort(lst):
    d = {}
    for el in lst:
        length = len(str(el))
        if length not in d.keys():
            d[length] = [el]
        else:
            d[length].append(el) 
    
    for k,v in d.items():
        sorted_list = sorted(v)
        d[k] = sorted_list
        
    od = collections.OrderedDict(sorted(d.items(), reverse=True))
  
    l = list(od.values())
    return [item for sublist in l for item in sublist]
        

print(digit_sort([77, 23, 5, 7, 101]))
print(digit_sort([1, 5, 9, 2, 789, 563, 444]))
print(digit_sort([53219, 3772, 564, 32, 1]))
```
