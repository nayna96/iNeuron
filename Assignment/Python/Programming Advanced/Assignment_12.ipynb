### 1. For this challenge, forget how to add two numbers together. The best explanation on what to do for this function is this meme:
    
**Examples**

meme_sum(26, 39) ➞ 515<br>
\# 2+3 = 5, 6+9 = 15<br>
\# 26 + 39 = 515

meme_sum(122, 81) ➞ 1103<br>
\# 1+0 = 1, 2+8 = 10, 2+1 = 3<br>
\# 122 + 81 = 1103

meme_sum(1222, 30277) ➞ 31499

**Ans:**

```python
def meme_sum(num1, num2):
    res = ""
    
    maxlen = max(len(str(num1)), len(str(num2)))

    n1 = str(num1).zfill(maxlen)
    n2 = str(num2).zfill(maxlen)
        
    for i in range(maxlen):
        res += str(int(n1[i]) + int(n2[i]))
        
    return res

print(meme_sum(26, 39))
print(meme_sum(122, 81))
print(meme_sum(1222, 30277))
```

### 2. Given an integer, create a function that returns the next prime. If the number is prime, return the number itself.

**Examples**

next_prime(12) ➞ 13<br>
next_prime(24) ➞ 29<br>
next_prime(11) ➞ 11

**Ans:**

```python
import math

def isPrime(n):
    if(n <= 1):
        return False

    if(n <= 3):
        return True

    if(n % 2 == 0 or n % 3 == 0):
        return False

    for i in range(5,int(math.sqrt(n) + 1), 6):
        if(n % i == 0 or n % (i + 2) == 0):
            return False

    return True
```

```python
def next_prime(N):
    if(isPrime(N) == True):
        return N
    
    if (N <= 1):
        return 2

    prime = N
    found = False

    while(not found):
        prime = prime + 1

        if(isPrime(prime) == True):
            found = True

    return prime

print(next_prime(12))
print(next_prime(24))
print(next_prime(11))
```

### 3. If a person traveled up a hill for 18mins at 20mph and then traveled back down the same path at 60mph then their average speed traveled was 30mph.

Write a function that returns the average speed traveled given an uphill time, uphill rate and a downhill rate. Uphill time is given in minutes. Return the rate as an integer (mph). No rounding is necessary.

**Examples**

ave_spd(18, 20, 60) ➞ 30<br>
ave_spd(30, 10, 30) ➞ 15<br>
ave_spd(30, 8, 24) ➞ 12

**Ans:**

```python
def ave_spd(time, uphill_speed, downhill_speed):
    dist = time * uphill_speed / 60 
    total_dist = 2 * dist
    down_time = dist / downhill_speed * 60
    total_time = down_time + time 
    return int(total_dist / total_time * 60)

print(ave_spd(18, 20, 60))
print(ave_spd(30, 10, 30))
print(ave_spd(30, 8, 24))
```

### 4. The Kempner Function, applied to a composite number, permits to find the smallest integer greater than zero whose factorial is exactly divided by the number.

kempner(6) ➞ 3<br>
1! = 1 % 6 > 0<br>
2! = 2 % 6 > 0<br>
3! = 6 % 6 === 0

kempner(10) ➞ 5<br>
1! = 1 % 10 > 0<br>
2! = 2 % 10 > 0<br>
3! = 6 % 10 > 0<br>
4! = 24 % 10 > 0<br>
5! = 120 % 10 === 0

A Kempner Function applied to a prime will always return the prime itself.

kempner(2) ➞ 2<br>
kempner(5) ➞ 5

Given an integer n, implement a Kempner Function.

**Examples**

kempner(6) ➞ 3<br>
kempner(10) ➞ 5<br>
kempner(2) ➞ 2<br>
kempner(5) ➞ 5

**Ans:**

```python
def fact(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * fact(n-1)

def kempner(num):
    for i in range(1,num+1):
        if fact(i) % num == 0:
            return i

print(kempner(6))
print(kempner(10))
print(kempner(2))
print(kempner(5))
```

### 5. You work in a factory, and your job is to take items from a conveyor belt and pack them into boxes. Each box can hold a maximum of 10 kgs. Given a list containing the weight (in kg) of each item, how many boxes would you need to pack all of the items?

**Examples**

boxes([2, 1, 2, 5, 4, 3, 6, 1, 1, 9, 3, 2]) ➞ 5

\# Box 1 = [2, 1, 2, 5] (10kg)<br>
\# Box 2 = [4, 3] (7kg)<br>
\# Box 3 = [6, 1, 1] (8kg)<br>
\# Box 4 = [9] (9kg)<br>
\# Box 5 = [3, 2] (5kg)

**Ans:**

```python
import math

def boxes(lst):
    count = sum_ = 0
    
    for i in lst:
        if (sum_ + i) <= 10:
            sum_ += i
        else:
            count += 1
            sum_ = 0
            sum_ += i
    
    return count + 1

print(boxes([2, 1, 2, 5, 4, 3, 6, 1, 1, 9, 3, 2]))
```
