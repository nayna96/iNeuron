### 1. Create a function based on the input and output. Look at the examples, there is a pattern.

**Examples**

secret("p.one.two.three") ➞ "\<p class='one two three'>\</p>"<br>
secret("p.one") ➞ "\<p class='one'>\</p>"<br>
secret("p.four.five") ➞ "\<p class='four five'>\</p>"

**Ans:**

```python
import re

def secret(inp):
    pattern = "(?<=.)\w+"
    classname = " ".join(re.findall(pattern, inp))
    return "<p class='{}'></p>".format(classname)

print(secret("p.one.two.three"))
print(secret("p.one"))
print(secret("p.four.five"))
```

### 2. Create a function which counts how many lone 1s appear in a given number. Lone means the number doesn't appear twice or more in a row.

**Examples**

count_lone_ones(101) ➞ 2<br>
count_lone_ones(1191) ➞ 1<br>
count_lone_ones(1111) ➞ 0<br>
count_lone_ones(462) ➞ 0

**Ans:**

```python
def count_lone_ones(num):
    count = 0
    
    n = str(num)
    for i in range(len(n)):
        if n[i] == "1":
            if i == 0:
                if n[i+1] != "1":
                    count += 1

            elif i == len(n) - 1:
                if n[i-1] != "1":
                    count += 1

            elif i > 0 and i < len(n) - 1:
                if n[i-1] != "1" and n[i+1] != "1":
                    count += 1
            
    return count

print(count_lone_ones(101))
print(count_lone_ones(1191))
print(count_lone_ones(1111))
print(count_lone_ones(462))
```

### 3. Write a method that accepts two integer parameters rows and cols. The output is a 2d array of numbers displayed in column-major order, meaning the numbers shown increase sequentially down each column and wrap to the top of the next column to the right once the bottom of the current column is reached.

**Examples**

printGrid(3, 6) ➞ [<br>
  [1, 4, 7, 10, 13, 16],<br>
  [2, 5, 8, 11, 14, 17],<br>
  [3, 6, 9, 12, 15, 18]<br>
]

printGrid(5, 3) ➞ [<br>
  [1, 6, 11],<br>
  [2, 7, 12],<br>
  [3, 8, 13],<br>
  [4, 9, 14],<br>
  [5, 10, 15]<br>
]

printGrid(4, 1) ➞ [<br>
  [1],<br>
  [2],<br>
  [3],<br>
  [4]<br>
]

**Ans:**

```python
def printGrid(r,c):
    grid = [list(range(i, (r*c)+1, r)) for i in range(1, r+1)]
    return ',\n'.join([str(lst) for lst in grid])
    
print(printGrid(3, 6))
print()
print(printGrid(5, 3)) 
print()
print(printGrid(4, 1)) 
```

### 4. Given a list of integers, return the smallest positive integer not present in the list.

Here is a representative example. Consider the list:<br>
[-2, 6, 4, 5, 7, -1, 7, 1, 3, 6, 6, -2, 9, 10, 2, 2]

After reordering, the list becomes:<br>
[-2, -2, -1, 1, 2, 2, 3, 4, 5, 6, 6, 6, 7, 7, 9, 10]

from which we see that the smallest missing positive integer is 8.

**Examples**

min_miss_pos([-2, 6, 4, 5, 7, -1, 1, 3, 6, -2, 9, 10, 2, 2]) ➞ 8<br>
\# After sorting, list becomes [-2, -2, -1, 1, 2, 2, 3, 4, 5, 6, 6, 7, 9, 10]<br>
\# So the smallest missing positive integer is 8

min_miss_pos([5, 9, -2, 0, 1, 3, 9, 3, 8, 9]) ➞ 2<br>
\# After sorting, list becomes [-2, 0, 1, 3, 3, 5, 8, 9, 9, 9]<br>
\# So the smallest missing positive integer is 2

min_miss_pos([0, 4, 4, -1, 9, 4, 5, 2, 10, 7, 6, 3, 10, 9]) ➞ 1<br>
\# After sorting, list becomes [-1, 0, 2, 3, 4, 4, 4, 5, 6, 7, 9, 9, 10, 10]<br>
\# So the smallest missing positive integer is 1

**Ans:**

```python
def min_miss_pos(lst):
    min_ = min(lst)    
    max_ = max(lst)

    l = [i for i in range(min_,max_+1) if i not in lst if i > 0]
    return min(l)
    
print(min_miss_pos([-2, 6, 4, 5, 7, -1, 1, 3, 6, -2, 9, 10, 2, 2]))
print(min_miss_pos([5, 9, -2, 0, 1, 3, 9, 3, 8, 9]))
print(min_miss_pos([0, 4, 4, -1, 9, 4, 5, 2, 10, 7, 6, 3, 10, 9]))
```

### 5. Google is launching a network of autonomous pizza delivery drones and wants you to create a flexible rewards system (Pizza Points™) that can be tweaked in the future. The rules are simple: if a customer has made at least N orders of at least Y price, they get a FREE pizza!

Create a function that takes a dictionary of customers, a minimum number of orders and a minimum order price. Return a list of customers that are eligible for a free pizza.

**Examples**

customers = {<br>
"Batman": [22, 30, 11, 17, 15, 52, 27, 12],<br>
"Spider-Man": [5, 17, 30, 33, 40, 22, 26, 10, 11, 45]<br>
}

pizza_points(customers, 5, 20) ➞ ["Spider-Man"]<br>
pizza_points(customers, 3, 10) ➞ ["Batman", "Spider-Man"]<br>
pizza_points(customers, 5, 100) ➞ []

**Ans:**

```python
customers = {
"Batman": [22, 30, 11, 17, 15, 52, 27, 12],
"Spider-Man": [5, 17, 30, 33, 40, 22, 26, 10, 11, 45]
}

def pizza_points(customers, N, P):
    winners = []
    for name,orders in customers.items():
        count = 0
        for order in orders:
            if order >= P:
                count += 1
                
        if count > N:
            winners.append(name)
            
    return winners

print(pizza_points(customers, 5, 20))
print(pizza_points(customers, 3, 10))
print(pizza_points(customers, 5, 100))
```
