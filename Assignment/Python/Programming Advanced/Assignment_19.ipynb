### 1. Create a checker board generator, which takes as inputs n and 2 elements to generate an n x n checkerboard with those two elements as alternating squares.

**Examples**

checker_board(2, 7, 6) ➞ [<br>
  [7, 6],<br>
  [6, 7]<br>
]

checker_board(3, "A", "B") ➞ [<br>
  ["A", "B", "A"],<br>
  ["B", "A", "B"],<br>
  ["A", "B", "A"]<br>
]

checker_board(4, "c", "d") ➞ [<br>
  ["c", "d", "c", "d"],<br>
  ["d", "c", "d", "c"],<br>
  ["c", "d", "c", "d"],<br>
  ["d", "c", "d", "c"]<br>
]

checker_board(4, "c", "c") ➞ "invalid"

**Ans:**

```python
def checker_board(n,a,b):
    if a == b:
        return "invalid"

    l1 = [[a,b][i%2] for i in range(n)]
    l2 = [[b,a][i%2] for i in range(n)]
    board = [l1 if i % 2 == 0 else l2 for i in range(n)]            
    
    return ',\n'.join([str(lst) for lst in board])
    
print(checker_board(2, 7, 6) )
print()
print(checker_board(3, "A", "B"))
print()
print(checker_board(4, "c", "d"))
print()
print(checker_board(4, "c", "c"))
```

### 2. A string is an almost-palindrome if, by changing only one character, you can make it a palindrome. Create a function that returns True if a string is an almost-palindrome and False otherwise.

**Examples**

almost_palindrome("abcdcbg") ➞ True<br>
\# Transformed to "abcdcba" by changing "g" to "a".

almost_palindrome("abccia") ➞ True<br>
\# Transformed to "abccba" by changing "i" to "b".

almost_palindrome("abcdaaa") ➞ False<br>
\# Can't be transformed to a palindrome in exactly 1 turn.

almost_palindrome("1234312") ➞ False

**Ans:**

```python
def almost_palindrome(string):
    rev = string[::-1]
    count = 0
    for i in range(len(string)):
        if string[i]==rev[i]:
            count = count + 1
    return count == len(string)-2
    
print(almost_palindrome("abcdcbg"))
print(almost_palindrome("abccia"))
print(almost_palindrome("abcdaaa"))
print(almost_palindrome("1234312"))
```

### 3. Create a function that finds how many prime numbers there are, up to the given integer.

**Examples**

prime_numbers(10) ➞ 4<br>
\# 2, 3, 5 and 7

prime_numbers(20) ➞ 8<br>
\# 2, 3, 5, 7, 11, 13, 17 and 19

prime_numbers(30) ➞ 10<br>
\# 2, 3, 5, 7, 11, 13, 17, 19, 23 and 29

**Ans:**

```python
def prime_numbers(n):
    count = 0
    for i in range(n):
        if i>1:
            for j in range(2,i):
                if(i % j==0):
                    break
            else:
                count += 1
    return count

print(prime_numbers(10))
print(prime_numbers(20))
print(prime_numbers(30))
```

### 4. If today was Monday, in two days, it would be Wednesday.

Create a function that takes in a list of days as input and the number of days to increment by. Return a list of days after n number of days has passed.

**Examples**

after_n_days(["Thursday", "Monday"], 4) ➞ ["Monday", "Friday"]<br>
after_n_days(["Sunday", "Sunday", "Sunday"], 1) ➞ ["Monday", "Monday", "Monday"]<br>
after_n_days(["Monday", "Tuesday", "Friday"], 1) ➞ ["Tuesday", "Wednesday", "Saturday"]

**Ans:**

```python
d = {0: "Monday", 1: "Tuesday", 2: "Wednesday", 3: "Thursday", 4: "Friday", 5: "Saturday", 6: "Sunday"}

def after_n_days(days, n):
    l = []
    for day in days:
        t = list(d.keys())[list(d.values()).index(day)] + n
        t = t-7 if t > 6 else t
        l.append(d[t])
        
    return l
            
print(after_n_days(["Thursday", "Monday"], 4))
print(after_n_days(["Sunday", "Sunday", "Sunday"], 1))
print(after_n_days(["Monday", "Tuesday", "Friday"], 1))
```

### 5. You are in the process of creating a chat application and want to add an anonymous name feature. This anonymous name feature will create an alias that consists of two capitalized words beginning with the same letter as the users first name.

Create a function that determines if the list of users is mapped to a list of anonymous names correctly.

**Examples**

is_correct_aliases(["Adrian M.", "Harriet S.", "Mandy T."], ["Amazing Artichoke", "Hopeful Hedgehog", "Marvelous Mouse"]) ➞ True<br>
is_correct_aliases(["Rachel F.", "Pam G.", "Fred Z.", "Nancy K."], ["Reassuring Rat", "Peaceful Panda", "Fantastic Frog", "Notable Nickel"]) ➞ True

is_correct_aliases(["Beth T."], ["Brandishing Mimosa"]) ➞ False<br>
\# Both words in "Brandishing Mimosa" should begin with a "B" - "Brandishing Beaver" would do the trick.

**Ans:**

```python
def is_correct_aliases(users, an_names):
    for i in range(len(users)):
        names = an_names[i].split()
        if users[i][0] != names[0][0] or users[i][0] != names[1][0]:
            return False
    return True

print(is_correct_aliases(["Adrian M.", "Harriet S.", "Mandy T."], ["Amazing Artichoke", "Hopeful Hedgehog", "Marvelous Mouse"]))
print(is_correct_aliases(["Rachel F.", "Pam G.", "Fred Z.", "Nancy K."], ["Reassuring Rat", "Peaceful Panda", "Fantastic Frog", "Notable Nickel"]))
print(is_correct_aliases(["Beth T."], ["Brandishing Mimosa"]))
```
