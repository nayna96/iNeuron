### 1. Given a sentence as txt, return True if any two adjacent words have this property: One word ends with a vowel, while the word immediately after begins with a vowel (a e i o u).

**Examples**

vowel_links("a very large appliance") ➞ True<br>
vowel_links("go to edabit") ➞ True<br>
vowel_links("an open fire") ➞ False<br>
vowel_links("a sudden applause") ➞ False

**Ans:**

```python
def vowel_links(string):
    vowels = "aAeEiIoOuu"
    words = string.split()
    for i in range(1, len(words)-1):
        if (words[i][-1] in vowels) and (words[i+1][0] in vowels):
            return True
    
    return False

print(vowel_links("a very large appliance"))
print(vowel_links("go to edabit"))
print(vowel_links("an open fire"))
print(vowel_links("a sudden applause"))
```

### 2. You are given three inputs: a string, one letter, and a second letter.

Write a function that returns True if every instance of the first letter occurs before every instance of the second letter.

**Examples**

first_before_second("a rabbit jumps joyfully", "a", "j") ➞ True<br>
\# Every instance of "a" occurs before every instance of "j".

first_before_second("knaves knew about waterfalls", "k", "w") ➞  True

first_before_second("happy birthday", "a", "y") ➞ False<br>
\# The "a" in "birthday" occurs after the "y" in "happy".

first_before_second("precarious kangaroos", "k", "a") ➞ False

**Ans:**

```python
def first_before_second(string, let1, let2):
    import re
    end = [m.start() for m in re.finditer(let1, string)][-1]
    
    for i in range(end):
        if string[i] == let2:
            return False
    
    return True

print(first_before_second("a rabbit jumps joyfully", "a", "j"))
print(first_before_second("knaves knew about waterfalls", "k", "w"))
print(first_before_second("happy birthday", "a", "y"))
print(first_before_second("precarious kangaroos", "k", "a"))
```

### 3. Create a function that returns the characters from a list or string r on odd or even positions, depending on the specifier s. The specifier will be "odd" for items on odd positions (1, 3, 5, ...) and "even" for items on even positions (2, 4, 6, ...).

**Examples**

char_at_pos([2, 4, 6, 8, 10], "even") ➞ [4, 8]<br>
\# 4 & 8 occupy the 2nd & 4th positions

char_at_pos("EDABIT", "odd") ➞ "EAI"<br>
\# "E", "A" and "I" occupy the 1st, 3rd and 5th positions

char_at_pos(["A", "R", "B", "I", "T", "R", "A", "R", "I", "L", "Y"], "odd") ➞ ["A", "B", "T", "A", "I", "Y"]

**Ans:**

```python
def char_at_pos(lst, ev_od):
    if ev_od == "even":
        res = [i for i in lst if lst.index(i) % 2 != 0]
        return ''.join(res) if type(lst) == str else res
    else:
        res = [i for i in lst if lst.index(i) % 2 == 0]
        return ''.join(res) if type(lst) == str else res

print(char_at_pos([2, 4, 6, 8, 10], "even"))
print(char_at_pos("EDABIT", "odd"))
print(char_at_pos(["A", "R", "B", "I", "T", "R", "A", "R", "I", "L", "Y"], "odd"))
```

### 4. Write a function that returns the greatest common divisor of all list elements. If the greatest common divisor is 1, return 1.

**Examples**

GCD([10, 20, 40]) ➞ 10<br>
GCD([1, 2, 3, 100]) ➞ 1<br>
GCD([1024, 192, 2048, 512]) ➞ 64

**Ans:**

```python
def find_gcd(x, y):
    while(y):
        x, y = y, x % y
    return x

def GCD(lst):
    num1=lst[0]
    num2=lst[1]
    
    gcd=find_gcd(num1,num2)
    
    for i in range(2,len(lst)):
        gcd=find_gcd(gcd,lst[i])

    return gcd

print(GCD([10, 20, 40]))
print(GCD([1, 2, 3, 100]))
print(GCD([1024, 192, 2048, 512]))
```

### 5. A number/string is a palindrome if the digits/characters are the same when read both forward and backward. Examples include "racecar" and 12321. Given a positive number n, check if n or the binary representation of n is palindromic. Return the following:

- "Decimal only." if only n is a palindrome.
- "Binary only." if only the binary representation of n is a palindrome.
- "Decimal and binary." if both are palindromes.
- "Neither!" if neither are palindromes.

**Examples**

palindrome_type(1306031) ➞ "Decimal only."<br>
\# decimal = 1306031<br>
\# binary  = "100111110110110101111"

palindrome_type(427787) ➞ "Binary only."<br>
\# decimal = 427787<br>
\# binary  = "1101000011100001011"

palindrome_type(313) ➞ "Decimal and binary."<br>
\# decimal = 313<br>
\# binary  = 100111001

palindrome_type(934) ➞ "Neither!"<br>
\# decimal = 934<br>
\# binary  = "1110100110"


**Ans:**

```python
def palindrome_type(decimal):
    binary = bin(decimal).replace("0b", "")
    d = True if str(decimal) == str(decimal)[::-1] else False
    b = True if str(binary) == str(binary)[::-1] else False
        
    if d == True:
        if b == True:
            return "Decimal and binary."
        else:
            return "Decimal only."
    else:
        if b == True:
            return "Binary only."
        else:
            return "Neither!"
        
print(palindrome_type(1306031))
print(palindrome_type(427787))
print(palindrome_type(313))
print(palindrome_type(934))
```
