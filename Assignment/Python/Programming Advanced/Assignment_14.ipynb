### 1. Given a list of numbers, create a function that removes 25% from every number in the list except the smallest number, and adds the total amount removed to the smallest number.

**Examples**

show_the_love([4, 1, 4]) ➞ [3, 3, 3]<br>
show_the_love([16, 10, 8]) ➞ [12, 7.5, 14.5]<br>
show_the_love([2, 100]) ➞ [27, 75]

**Ans:**

```python
def show_the_love(lst):
    sum_ = 0
    index = lst.index(min(lst))
    
    for i in range(len(lst)):
        if lst[i] != min(lst):
            n = 0.25 * lst[i]
            sum_ += n
            lst[i] -= n
            
    lst[index] += sum_
    return lst

print(show_the_love([4, 1, 4]))
print(show_the_love([16, 10, 8]))
print(show_the_love([2, 100]))
```

### 2. Create a function that takes in two words as input and returns a list of three elements, in the following order:

1. Shared letters between two words.
2. Letters unique to word 1.
3. Letters unique to word 2.

Each element should have unique letters, and have each letter be alphabetically sorted.

**Examples**

letters("sharp", "soap") ➞ ["aps", "hr", "o"]<br>
letters("board", "bored") ➞ ["bdor", "a", "e"]<br>
letters("happiness", "envelope") ➞ ["enp", "ahis", "lov"]

letters("kerfuffle", "fluffy") ➞ ["flu", "ekr", "y"]<br>
\# Even with multiple matching letters (e.g. 3 f's), there should<br> 
\# only exist a single "f" in your first element.

letters("match", "ham") ➞ ["ahm", "ct", ""]<br>
\# "ham" does not contain any letters that are not found already<br> 
\# in "match".

**Ans:**

```python
def letters(str1, str2):
    lst = []
    
    s1 = set(str1)
    s2 = set(str2)
    
    lst.append("".join(sorted(s1.intersection(s2))))
    lst.append("".join(sorted(s1 -s2)))
    lst.append("".join(sorted(s2 - s1)))
    
    return lst

print(letters("sharp", "soap"))
print(letters("board", "bored"))
print(letters("happiness", "envelope"))
print(letters("kerfuffle", "fluffy"))
print(letters("match", "ham"))
```

### 3. Write a function that pairs the first number in an array with the last, the second number with the second to last, etc.

**Examples**

pairs([1, 2, 3, 4, 5, 6, 7]) ➞ [[1, 7], [2, 6], [3, 5], [4, 4]]<br>
pairs([1, 2, 3, 4, 5, 6]) ➞ [[1, 6], [2, 5], [3, 4]]<br>
pairs([5, 9, 8, 1, 2]) ➞ [[5, 2], [9, 1], [8, 8]]<br>
pairs([]) ➞ []

**Ans:**

```python
def pairs(lst):
    return [[lst[i], lst[len(lst) - i - 1]] for i in range(len(lst)) if i <len(lst)/2]

print(pairs([1, 2, 3, 4, 5, 6, 7]))
print(pairs([1, 2, 3, 4, 5, 6]))
print(pairs([5, 9, 8, 1, 2]))
print(pairs([]))
```

### 4. Write a function that adds two numbers. The catch, however, is that the numbers will be strings.

**Examples**

add_str_nums("4", "5") ➞ "9"<br>
add_str_nums("abcdefg", "3") ➞ "-1"<br>
add_str_nums("1", "") ➞ "1"<br>
add_str_nums("1874682736267235927359283579235789257", "32652983572985729") ➞ "1874682736267235927391936562808774986"

**Ans:**

```python
def add_str_nums(n1, n2):
    if n1 == "":
        n1 = "0"
    if n2 == "":
        n2 = "0"
    if n1.isnumeric() and n2.isnumeric():
        return int(n1) + int(n2)
    else:
        return -1

print(add_str_nums("4", "5"))
print(add_str_nums("abcdefg", "3"))
print(add_str_nums("1", ""))
print(add_str_nums("1874682736267235927359283579235789257", "32652983572985729"))
```

### 5. lPaeesh le pemu mnxit ehess rtnisg! Oh, sorry, that was supposed to say: Please help me unmix these strings!

Somehow my strings have all become mixed up; every pair of characters has been swapped. Help me undo this so I can understand my strings again.

**Examples**

unmix("123456") ➞ "214365"<br>
unmix("hTsii  s aimex dpus rtni.g") ➞ "This is a mixed up string."<br>
unmix("badce") ➞ "abcde"

**Ans:**

```python
def unmix(string):
    return ''.join(string[i:i+2][::-1] for i in range(0,len(string),2))

print(unmix("123456"))
print(unmix("hTsii  s aimex dpus rtni.g"))
print(unmix("badce"))
```
