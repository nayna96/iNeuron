### 1. Rondo Form is a type of musical structure, in which there is a recurring theme/refrain, notated as A. Here are the rules for valid rondo forms:

\- Rondo forms always start and end with an A section.<br>
\- In between the A sections, there should be contrasting sections notated as B, then C, then D, etc... No letter should be skipped.<br>
\- There shouldn't be any repeats in the sequence (such as ABBACCA).

Create a function which validates whether a given string is a valid Rondo Form.

**Examples**

valid_rondo("ABACADAEAFAGAHAIAJA") ➞ True<br>
valid_rondo("ABA") ➞ True<br>
valid_rondo("ABBACCA") ➞ False<br>
valid_rondo("ACAC") ➞ False<br>
valid_rondo("A") ➞ False

**Ans:**

```python
import re

def valid_rondo(string):
    return len(string) > 2 and string[0] == string[-1] == 'A' and all(i * 2 not in string for i in string)

print(valid_rondo("ABACADAEAFAGAHAIAJA"))
print(valid_rondo("ABA"))
print(valid_rondo("ABBACCA"))
print(valid_rondo("ACAC"))
print(valid_rondo("A"))
```

### 2. Create a function that returns the whole of the first sentence which contains a specific word. Include the full stop at the end of the sentence.

**Examples**

txt = "I have a cat. I have a mat. Things are going swell."

sentence_searcher(txt, "have") ➞ "I have a cat."<br>
sentence_searcher(txt, "MAT") ➞ "I have a mat."<br>
sentence_searcher(txt, "things") ➞ "Things are going swell."<br>
sentence_searcher(txt, "flat") ➞ ""

**Ans:**

```python
txt = "I have a cat. I have a mat. Things are going swell."

def sentence_searcher(txt, word):
    txt = txt.split('.')
    for i in txt:
        if word.lower() in i.lower():
            return i.strip()+'.'
    return ""

print(sentence_searcher(txt, "have"))
print(sentence_searcher(txt, "MAT"))
print(sentence_searcher(txt, "things"))
print(sentence_searcher(txt, "flat"))
```

### 3. Given a number, find the "round "of each digit of the number. An integer is called "round" if all its digits except the leftmost (most significant) are equal to zero.

\- Round numbers: 4000, 1, 9, 800, 90<br>
\- Not round numbers: 110, 707, 222, 1001

Create a function that takes a number and returns the "round" of each digit (except if the digit is zero) as a string. Check out the following examples for more clarification.

**Examples**

sum_round(101) ➞ "1 100"<br>
sum_round(1234) ➞ "4 30 200 1000"<br>
sum_round(54210) ➞ "10 200 4000 50000"

**Ans:**

```python
def sum_round(num):
    l = []
    n = str(num)[::-1]
    for i in range(len(n)):
        if n[i] != "0":
            l.append(n[i] + (i * "0"))
    return " ".join(l)

print(sum_round(101))
print(sum_round(1234))
print(sum_round(54210))
```

### 4. Your task, is to create N x N multiplication table, of size n provided in parameter.

For example, when n is 5, the multiplication table is:<br>
\- 1, 2, 3, 4, 5<br>
\- 2, 4, 6, 8, 10<br>
\- 3, 6, 9, 12, 15<br>
\- 4, 8, 12, 16, 20<br>
\- 5, 10, 15, 20, 25

This example will result in:<br>
[[1, 2, 3, 4, 5], [2, 4, 6, 8, 10], [3, 6, 9, 12, 15], [4, 8, 12, 16, 20], [5, 10, 15, 20, 25]]

**Examples**

multiplication_table(1) ➞ [[1]]<br>
multiplication_table(3) ➞ [[1, 2, 3], [2, 4, 6], [3, 6, 9]]

**Ans:**

```python
def multiplication_table(num):
    return [[i * j for j in range(1, num+1)] for i in range(1, num+1)]
    
print(multiplication_table(5))
print(multiplication_table(1))
print(multiplication_table(3))
```

### 5. Create a function that returns True if two lines rhyme and False otherwise. For the purposes of this exercise, two lines rhyme if the last word from each sentence contains the same vowels.

**Examples**

does_rhyme("Sam I am!", "Green eggs and ham.") ➞ True

does_rhyme("Sam I am!", "Green eggs and HAM.") ➞ True<br>
\# Capitalization and punctuation should not matter.

does_rhyme("You are off to the races", "a splendid day.") ➞ False<br>
does_rhyme("and frequently do?", "you gotta move.") ➞ False

**Ans:**

```python
import re

def does_rhyme(sent1, sent2):
    vowels = "aAeEiIoOuU"
    
    word1 = sent1.split()[-1].lower()    
    word2 = sent2.split()[-1].lower()
    
    vowels1 = [el for el in word1 if el in vowels]
    vowels2 = [el for el in word2 if el in vowels]
    
    if set(vowels1) == set(vowels2):
        return True
    else: 
        return False

print(does_rhyme("Sam I am!", "Green eggs and ham."))
print(does_rhyme("Sam I am!", "Green eggs and HAM."))
print(does_rhyme("You are off to the races", "a splendid day."))
print(does_rhyme("and frequently do?", "you gotta move."))
```
