### 1. Create a function that takes the width, height and character and returns a picture frame as a 2D list.

**Examples**

get_frame(4, 5, "#") ➞ [<br>
  ["####"],<br>
  ["#  #"],<br>
  ["#  #"],<br>
  ["#  #"],<br>
  ["####"]<br>
]<br>
\# Frame is 4 characters wide and 5 characters tall.

get_frame(10, 3, "*") ➞ [<br>
  ["\*\*\*\*\*\*\*\*\*\*"],<br>
  ["*        *"],<br>
  ["\*\*\*\*\*\*\*\*\*\*"]<br>
]<br>
\# Frame is 10 characters and wide and 3 characters tall.

get_frame(2, 5, "0") ➞ "invalid"<br>
\# Frame's width is not more than 2.

**Ans:**

```python
def get_frame(w, h, ch):
    if w != 2:
        l = []

        l.append([ch * w])

        for i in range(1, h -1):
            l.append([ch + (' ' * (w-2)) + ch])

        l.append([ch * w])

        for item in l:
            print(' '.join(str(x) for x in item))
    else:
        print("invalid")
```

```python
get_frame(4, 5, "#")
print()
get_frame(10, 3, "*")
print()
get_frame(2, 5, "0") 
```

### 2. Write three functions:
  1. boolean_and
  2. boolean_or
  3. boolean_xor
  
These functions should evaluate a list of True and False values, starting from the leftmost element and evaluating pairwise.

**Examples**

boolean_and([True, True, False, True]) ➞ False<br>
\# [True, True, False, True] => [True, False, True] => [False, True] => False

boolean_or([True, True, False, False]) ➞ True<br>
\# [True, True, False, True] => [True, False, False] => [True, False] => True

boolean_xor([True, True, False, False]) ➞ False<br>
\# [True, True, False, False] => [False, False, False] => [False, False] => False

**Ans:**

```python
import operator
from functools import reduce

def boolean_and(lst):
    return reduce(operator.and_, lst)

def boolean_or(lst):
    return reduce(operator.or_, lst)

def boolean_xor(lst):
    return reduce(operator.xor, lst)

print(boolean_and([True, True, False, True]))
print(boolean_or([True, True, False, False]))
print(boolean_xor([True, True, False, False]))
```

### 3. Create a function that creates a box based on dimension n.

**Examples**

make_box(5) ➞ [<br>
  "#####",<br>
  "#   #",<br>
  "#   #",<br>
  "#   #",<br>
  "#####"<br>
]

make_box(3) ➞ [<br>
  "###",<br>
  "# #",<br>
  "###"<br>
]

make_box(2) ➞ [<br>
  "##",<br>
  "##"<br>
]

make_box(1) ➞ [<br>
  "#"<br>
]

**Ans:**

```python
def make_box(n):
    if n == 1:
        print ("#")
    else:
        l = []

        l.append(["#" * n])

        for i in range(1, n -1):
            l.append(["#" + (' ' * (n-2)) + "#"])

        l.append(["#" * n])

        for item in l:
            print(' '.join(str(x) for x in item)) 
```

```python
make_box(5) 
print()
make_box(3) 
print()
make_box(2)
print()
make_box(1)
```

### 4. Given a common phrase, return False if any individual word in the phrase contains duplicate letters. Return True otherwise.

**Examples**

no_duplicate_letters("Fortune favours the bold.") ➞ True<br>
no_duplicate_letters("You can lead a horse to water, but you can't make him drink.") ➞ True

no_duplicate_letters("Look before you leap.") ➞ False<br>
\# Duplicate letters in "Look" and "before".

no_duplicate_letters("An apple a day keeps the doctor away.") ➞ False<br>
\# Duplicate letters in "apple", "keeps", "doctor", and "away".<br>

**Ans:**

```python
def no_duplicate_letters(string):
    words = string.split()
    
    for word in words:
        if len(set(word)) != len(word):
            return False
    
    return True

print(no_duplicate_letters("Fortune favours the bold."))
print(no_duplicate_letters("You can lead a horse to water, but you can't make him drink."))
print(no_duplicate_letters("Look before you leap."))
print(no_duplicate_letters("An apple a day keeps the doctor away."))
```

### 5. Write a regular expression that will match the states that voted yes to President Trump's impeachment. You must use RegEx positive lookahead.

**Example**

txt = "Texas = no, California = yes, Florida = yes, Michigan = no"<br>
pattern = "yourregularexpressionhere"

re.findall(pattern, txt) ➞ ["California", "Florida"]

**Ans:**

```python
import re

txt = "Texas = no, California = yes, Florida = yes, Michigan = no"
pattern = "\w+(?= = yes)"

print(re.findall(pattern, txt))
```
