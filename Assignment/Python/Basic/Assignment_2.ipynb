### 11. How can you tell the difference between break and continue?

**Ans:**

When break keyword is encountered, it will exit the loop. 

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

In case of continue keyword, the current iteration that is running will be stopped, and it will proceed with the next iteration

```python
for i in range(10):
    if i == 5:
        continue
    print(i)
```

### 12. In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?

**Ans:**

All the three range functions will give the same output with for loop

```python
for i in range(10):
    print(i)
```

```python
for i in range(0, 10):
    print(i)
```

```python
for i in range(0, 10, 1):
    print(i)
```

range(n) will always start from 0 and end in n-1 with default step size 1

```python
for i in range(10):
    print(i)
```

In range (m,n) we can specify both the start and end point with default step size as 1 

```python
for i in range(3,10):
    print(i)
```

In range(m,n,p) we can also specify the step size along with the start and end point. 

```python
for i in range(3,10,2):
    print(i)
```
