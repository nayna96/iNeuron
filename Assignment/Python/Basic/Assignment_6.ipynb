### 1. What are escape characters, and how do you use them?

**Ans**

Escape characters are used to insert characters that are illegal in a string.

We use an escape character by using a backslash \ followed by the character we want to insert.

**Example:**

```python
txt = "We are the so-called \"Vikings\" from the north."
txt
```

Some other escape characters are-

\' -> Single Quote<br>
\\ -> Backslash<br>
\n -> New Line<br>
\r -> Carriage Return<br>
\t -> Tab

### 2. What do the escape characters n and t stand for?

**Ans**

\n stands for new line

```python
text = "ABCD\nEFGH\nIJKL"
print(text)
```

\t stands for tab space

```python
text = "ABCD\tEFGH\tIJKL"
print(text)
```

### 9. What are the methods for right-justifying, left-justifying, and centering a string?

**Ans**

**str.rjust**, **str.ljust** and **str.center** are the methods for right-justifying, left-justifying, and centering a string respectively.

They return a string that is at least width characters wide, created by padding the string s with the character fillchar (default is a space) until the given width on the right, left or both sides. The string is never truncated.

```python
text = "This is a text"
text
```

**Right-justify**

```python
print (text.rjust(40)) 
print (text.rjust(40, '-')) 
```

**Left-justify**

```python
print (text.ljust(40)) 
print (text.ljust(40, '-')) 
```

**Center**

```python
print (text.center(40)) 
print (text.center(40, '#')) 
```

### 10. What is the best way to remove whitespace characters from the start or end?

**Ans**

```python
name = '  NA  '
len(name)
```

lstrip() -> lstrip() method removes whitespace at the beginning of a string.

```python
print (name.lstrip())
print(len(name.lstrip()))
```

rstrip() -> rstrip() method removes whitespace at the end of a string.

```python
print (name.rstrip())
print(len(name.rstrip()))
```

strip() -> strip() method removes whitespace at the beginning and end (both sides) of a string.

```python
print (name.strip())
print(len(name.strip()))
```
