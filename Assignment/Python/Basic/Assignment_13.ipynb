### 2. What do you pass to csv.reader() and csv.writer() to create reader and writer objects?

**Ans:**

We pass csv file object to csv.reader() and csv.writer() to create reader and writer objects.

**csv.writer()**

```python
import csv

mydict =[{'title': 'The Weirdstone of Brisingamen', 'author': 'Alan Garner', 'year': '1960'},
        {'title': 'Perdido Street Station', 'author': 'China Miéville', 'year': '2000'},
        {'title': 'Thud!', 'author': 'Terry Pratchett', 'year': '2005'},
        {'title': 'The Spellman Files', 'author': 'Lisa Lutz', 'year': '2007'},
        {'title': 'Small Gods', 'author': 'Terry Pratchett', 'year': '1992'}]

fields = ['title', 'author', 'year']

filename = "books.csv"

with open(filename, 'w') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames = fields)
    writer.writeheader()
    writer.writerows(mydict)
```

**csv.reader()**

```python
filename = "books.csv"

fields = []
rows = []

with open(filename, 'r') as csvfile:
    csvreader = csv.reader(csvfile)
    fields = next(csvreader)
    for row in csvreader:
        rows.append(row)

    print("Total no. of rows: {}".format(csvreader.line_num))

print('Field names are:' + ', '.join(field for field in fields))
print()

print('First 5 rows are:')
print()
for row in rows:
    for col in row:
        print(col, end="\t\t")
    print()
```

### 4. What method takes a list argument and writes it to a CSV file?

**Ans:**

writerows takes a list argument and writes it to CSV file

```python
l = [
        ["a", "b", "c"],
        ["d", "e", "f"],
        ["g", "h", "i"]]

with open('test.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(l)
```

```python
with open('test.csv') as csvfile:
    reader = csv.reader(csvfile, delimiter='\t')
    for row in reader:
        print(', '.join(row))
```

### 5. What do the keyword arguments delimiter and line terminator do?

**Ans:**

The delimiter is the character that appears between cells on a row. By default, the delimiter for a CSV file is a comma.

The line terminator is the character that comes at the end of a row. By default, the line terminator is a newline.

```python
csvFile = open('example.tsv', 'w', newline='')
csvWriter = csv.writer(csvFile, delimiter='\t', lineterminator='\n\n')

csvWriter.writerow(['a', 'b', 'c'])
csvWriter.writerow(['d', 'e', 'f'])
csvWriter.writerow(['g', 'h', 'i'])

csvFile.close()
```

Passing delimeter='\t' and lineterminator='\n\n' changes the character between cells to a tab and the character between rows to two newlines.

```python
with open('example.tsv') as csvfile:
    reader = csv.reader(csvfile, delimiter='\t')
    for row in reader:
        print(', '.join(row))
```

### 6. What function takes a string of JSON data and returns a Python data structure?

**Ans:**

loads() takes a string of JSON data and returns a Python data structure

```python
import json

x = '{ "name":"A", "age":20, "city":"N"}'
y = json.loads(x)

print(y["age"])
print(type(x))
print(type(y))
```

### 7. What function takes a Python data structure and returns a string of JSON data?

**Ans:**

dumps() takes a string of JSON data and returns a Python data structure

```python
import json

x = {
  "name": "A",
  "age": 20,
  "city": "N"
}

y = json.dumps(x)

print(y)
print(type(x))
print(type(y))
```
