### 1. What is the result of the code, and why?
```python
def func(a, b=6, c=8):
    print(a, b, c)

func(1, 2)
```

**Ans**

````python
def func(a, b=6, c=8):
    print(a, b, c)
    
func(1, 2)
````

a = 1 -> We are passing a as 1 during function call<br>
b = 2 -> We have a default value as 6 but the value passed during function call overrides the default value. <br>
c = 8 -> When we do not pass any value during function call, it takes default value

### 3. How about this code: what is its result, and why?
```python
def func(a, *pargs):
    print(a, pargs)

func(1, 2, 3)
```

**Ans**

````python
def func(a, *pargs):
    print(a, pargs)

func(1, 2, 3)
````

- Here, \*pargs denotes non-keyword Arguments
- With \*pargs, we can pass any number of arguments to a function. 
- It returns a tuple of all the values that we pass to it.
- They are typically optional. 
- Leaving them out will return an empty tuple

kargs is also not a keyword, we can use any name instead of it. 

**Example:**

````python
def func(a, *xyz):
    print(type(xyz))
    print(a, xyz)

func(1, 2, 3)
````

### 4. What does this code print, and why?
```python
def func(a, **kargs):
    print(a, kargs)

func(a=1, c=3, b=2)
```

**Ans**

````python
def func(a, **kargs):
    print(a, kargs)

func(a=1, c=3, b=2)
````

- Here, \**kargs denotes keyword arguments. 
- Like \*pargs, with \**kargs also we can pass any number of arguments to a function. 
- It returns dictionary that maps each keyword to the value that we pass alongside it.
- They are also optional. 
- Leaving them out will return an empty dictionary

kargs is also not a keyword, we can use any name instead of it. 

**Example:**

````python
def func(a, **abc):
    print(type(abc))
    print(a, abc)

func(a=1, c=3, b=2)
````

### 5. What gets printed by this, and explain?
```python
def func(a, b, c=8, d=5): print(a, b, c, d)

func(1, *(5, 6))
```

**Ans**

````python
def func(a, b, c=8, d=5): print(a, b, c, d)

func(1, *(5, 6))
````

a = 1 -> 1 was passed for a 

\*(5,6) is a non-key word argument. For non-key word arguments, order matter. As we passed 5 and 6, b and c were replaced with 5 and 6 respectively.  

d = 5 -> 
d took default value of 5 as only 2 values were passed in non-keyword argument \*(). If 3rd value was passed in \*(), that value would have replaced d. 
Or if we would have passed a value for d like (1, \*(5,6), var_3) with a value for var_3, it would have replaced d. 

### 6. what is the result of this, and explain?
```python
def func(a, b, c): a = 2; b[0] = 'x'; c['a'] = 'y'

l=1; m=[1]; n={'a':0}

func(l, m, n)

l, m, n
```

**Ans**

````python
def func(a, b, c): a = 2; b[0] = 'x'; c['a'] = 'y'

l=1; m=[1]; n={'a':0}

func(l, m, n)

l, m, n
````

l = 1 -> 1 was passed for l<br>
m = ['x'] -> m[0] was 1, it was replaced by 'x'<br>
n = {'a': 'y'} -> n['a'] was 0, it was replaced by y
