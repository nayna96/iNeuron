### 1. How do you distinguish between shutil.copy() and shutil.copytree()?

**Ans**

The shutil.copy() function will copy a single file, while shutil.copytree() will copy an entire folder, along with all its contents.

**shutil.copy()**

```python
import os 
import shutil 

path = 'C:/Users/agarw/Desktop/CopyTest/A'
print("Before copying file:") 
print(os.listdir(path)) 

source = 'C:/Users/agarw/Desktop/CopyTest/A/test.txt'
destination = "C:/Users/agarw/Desktop/CopyTest/A/file(copy).txt"
dest = shutil.copy(source, destination) 

print("After copying file:") 
print(os.listdir(path)) 
```

**shutil.copytree()**

```python
import os 
import shutil 

path = 'C:/Users/agarw/Desktop/CopyTest'
print("Before copying file:") 
print(os.listdir(path)) 

src = 'C:/Users/agarw/Desktop/CopyTest/A'
dest = 'C:/Users/agarw/Desktop/CopyTest/B'

print(os.listdir(src))

destination = shutil.copytree(src, dest) 

print("\nAfter copying file:") 
print(os.listdir(path)) 

print(os.listdir(destination)) 
```

### 4. ZipFile objects have a close() function that is similar to that of File objects. What ZipFile method is similar to the open() method of File objects?

**Ans**

The zipfile.ZipFile() function is equivalent to the open() function; the first argument is the filename, and the second argument is the mode to open the ZIP file in (read, write, or append).

### 5. Create a programme that searches a folder tree for files with a certain file extension (such as .pdf or .jpg). Copy these files to a new folder from wherever they are now.

**Ans**

```python
import os 
import shutil


for folder, subfolders, files in os.walk('C:/Users/agarw/Desktop/CopyTest/A'): 

    for file in files:
        if file.endswith('.jpg'): 
            path = folder + "/" + file
            print("Current path:", path)
            dest = 'C:/Users/agarw/Desktop/CopyTest/B/' + file
            print(dest)
            
            shutil.copy2(path, dest) 
```
