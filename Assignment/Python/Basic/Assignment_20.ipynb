### 3. Create a CSV file called books.csv by using these lines:
title,author,year<br>
The Weirdstone of Brisingamen,Alan Garner,1960<br>
Perdido Street Station,China Miéville,2000<br>
Thud!,Terry Pratchett,2005<br>
The Spellman Files,Lisa Lutz,2007<br>
Small Gods,Terry Pratchett,1992

**Ans:**

```python
import csv
import pandas as pd

with open('books.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["title","author","year"])
    writer.writerow(["The Weirdstone of Brisingamen","Alan Garner","1960"])
    writer.writerow(["Perdido Street Station","China Miéville","2000"])
    writer.writerow(["Thud!","Terry Pratchett","2005"])
    writer.writerow(["The Spellman Files","Lisa Lutz","2007"])
    writer.writerow(["Small Gods","Terry Pratchett","1992"])
    
pd.read_csv('books.csv', sep=',', encoding='latin-1')
```

### 4. Use the sqlite3 module to create a SQLite database called books.db, and a table called books with these fields: title (text), author (text), and year (integer).

**Ans:**

```python
import sqlite3

c = sqlite3.connect("books.db")

c.execute("CREATE TABLE books(title text, author text, year int)")
```

### 5. Read books.csv and insert its data into the book table.

**Ans:**

```python
with open('books.csv','r') as file: 
    dr = csv.DictReader(file) 
    to_db = [(i['title'], i['author'], i['year']) for i in dr]

c.executemany("INSERT INTO books (title, author, year) VALUES (?, ?, ?);", to_db)
c.commit()
```

### 6. Select and print the title column from the book table in alphabetical order.

**Ans:**

```python
for i in c.execute("SELECT title FROM books ORDER BY title"):
    print(i)
```

### 7. From the book table, select and print all columns in the order of publication.

**Ans:**

```python
for i in c.execute("SELECT * FROM books ORDER BY year"):
    print(i)
```

### 8. Use the sqlalchemy module to connect to the sqlite3 database books.db that you just made in exercise 6.

**Ans:**

```python
from sqlalchemy import create_engine
db=create_engine('sqlite:////books.db')
pd.read_sql('select * from books',db)
```
