### 10. What is the error caused by this expression? What would you do about it? <br>
'I have eaten ' + 99 + ' burritos.'

**Ans:** 

The error caused by this expression is:

```python
'I have eaten ' + 99 + ' burritos.'
```

It gives this error because 99 is an integer. And we cannot concatenate int with str. 

To fix this error, we have to covert the integer to string by using str() method.

```python
'I have eaten ' + str(99) + ' burritos.'
```

Now we can print the statement.
