### 8. What are the three “mode” arguments that the open() function accepts?

**Ans**

The open() function accepts three modes - 'r' for read mode, 'w' for write mode, and 'a' for append mode.

**read mode**

```python
file = open('test.txt', 'r') 
for line in file: 
    print(line) 
```

**write mode**

```python
file = open('test.txt','w')

file.write("This is the write command\n") 
file.write("It allows us to write in a particular file\n") 

fileread = file = open('test.txt', 'r') 
for line in fileread: 
    print(line) 
```

**append mode**

```python
file = open('test.txt','a') 
file.write("This will add this line") 
file.close() 

fileread = file = open('test.txt', 'r') 
for line in fileread: 
    print(line) 
```

### 9. What happens if you open existing file in write mode?

**Ans**

An existing file opened in write mode is erased and completely overwritten.

```python
print("Existing file contents:\n ")
file = open('test.txt', 'r') 
for line in file: 
    print(line) 
    
file = open('test.txt','w')
file.write("Hello\n") 
file.write("world") 

print("\nAfter opening in write mode: \n")
file = open('test.txt', 'r') 
for line in file: 
    print(line) 
```

### 10. How do you tell the difference between read() and readlines()?

**Ans**

**read()**

The read() method returns the file's entire contents as a single string value. 

```python
f = open("test.txt", "r")

print(f.read())
```

**readlines()**

The readlines() method returns a list of strings, where each string is a line from the file's contents.

```python
f = open("test.txt", "r")

print(f.readlines())
```

### 11. How does a shelf value resemble a data structure?

**Ans**

A shelf value resembles a dictionary value; it has keys and values, along with key() and values() methods that work similarly to the dictionary methods of the same names.

```python
import shelve
s = shelve.open("test")
s['name'] = "NA"
s['age'] = 23
s['marks'] = 75
```

```python
print(s)
```

```python
list(s.items())
```

```python
list(s.keys())
```

```python
list(s.values())
```

```python
s.close()
```
