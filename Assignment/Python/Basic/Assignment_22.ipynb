### 4. What output does this code produce? Why?

```python

    global X
    X = 'NI'

func()
print(X)
```

**Ans:**

````python
X = 'iNeuron'
def func():
    global X
    X = 'NI'

func()
print(X)
````

Here, the global keyword is used inside the function. Hence, the global X value is also changed

### 6. How about this code: what is its output in Python 3, and explain?

```python
def func():
    X = 'NI'
    
    def nested():
        nonlocal X
        X = 'Spam'
    
    nested()
    print(X)

func()
```

**Ans:**

````python
def func():
    X = 'NI'
    
    def nested():
        nonlocal X
        X = 'Spam'
    
    nested()
    print(X)

func()
````

nested function assigns the value Spam to X. Hence, X is printed
