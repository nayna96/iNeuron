### Q4. What is the relationship between indexing and slicing?

**Ans:**

**Indexing**

Lists elements can be accessed using a numerical index in square brackets. List elements can also be accessed using a negative list index, which counts from the end of the list

```python
l = ['a', 'b', 'c', 'd', 'e', 'f']

print(l[-1])
print(l[-2])
print(l[-5])
print(l[-6])
print(l[-len(l)])
```

**Slicing**

Slicing is indexing syntax that extracts a portion from a list. 

If a is a list, then a[m:n] returns the portion of a:
- Starting with postion m
- Up to but not including n
- Negative indexing can also be used

```python
l = ['a', 'b', 'c', 'd', 'e', 'f']

print(l[2:5])
print(l[-5:-2])
print(l[1:4])
print(l[-5:-2] == l[1:4])
```

### Q8. What is the benefit of first checking the target string with in or not in before using the index method to find a substring?

**Ans:**

```python
string = "Hello world"
search = "abc"

x= string.index(search)
print(x)
```

If we check the target string with "in" or "not in" before using the index method to find a substring, we will not get a Value Errror if the substring is not present

```python
string = "Hello world"
search = "abc"

if search in string:
    x= string.index(search)
    print(x)
else:
    print("String not found")
```
