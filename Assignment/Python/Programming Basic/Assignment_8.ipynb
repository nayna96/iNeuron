### 1. Write a Python Program to Add Two Matrices?

**Ans:**

```python
mat1 = np.mat(([1,2], [3,4]))
mat2 = np.mat(([5,6], [7,8]))

mat1 + mat2
```

### 2. Write a Python Program to Multiply Two Matrices?

**Ans:**

```python
mat1 = np.mat(([1,2], [3,4]))
mat2 = np.mat(([5,6], [7,8]))

mat1 * mat2
```

### 3. Write a Python Program to Transpose a Matrix?

**Ans:**

```python
mat = np.mat(([1,2], [3,4]))
print(mat)
print("\nTranspose")
print(mat.T)
```

### 4. Write a Python Program to Sort Words in Alphabetic Order?

**Ans:**

```python
string = "hello world. this is a text"

words = string.split()

words.sort()

for word in words:
    print(word)
```

### 5. Write a Python Program to Remove Punctuation From a String?

**Ans:**

```python
import re

string = "Hello, world : for ! Python ;"
newtext = re.sub(r'[^A-Za-z]+', ' ', string)
newtext
```
