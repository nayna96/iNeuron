### 1. Create a function that takes a string and returns a string in which each character is repeated once.

**Examples**

double_char("String") ➞ "SSttrriinngg"<br>
double_char("Hello World!") ➞ "HHeelllloo  WWoorrlldd!!"<br>
double_char("1234!_ ") ➞ "11223344!!__  "

**Ans:**

```python
def double_char(string):
    l = []
    for i in string:
        l.append(i)
        l.append(i)
    return "".join(l)

print(double_char("String"))
print(double_char("Hello World!"))
print(double_char("1234!_ "))
```

### 2. Create a function that reverses a boolean value and returns the string "boolean expected" if another variable type is given.

**Examples**

reverse(True) ➞ False<br>
reverse(False) ➞ True<br>
reverse(0) ➞ "boolean expected"<br>
reverse(None) ➞ "boolean expected"

**Ans:**

```python
def reverse(bool_):
    if type(bool_) == bool:
        return not bool_
    else:
        return "boolean expected"
    
print(reverse(True))
print(reverse(False))
print(reverse(0))
print(reverse(None) )
```

### 3. Create a function that returns the thickness (in meters) of a piece of paper after folding it n number of times. The paper starts off with a thickness of 0.5mm.

**Examples**

num_layers(1) ➞ "0.001m"<br>
Paper folded once is 1mm (equal to 0.001m)

num_layers(4) ➞ "0.008m"<br>
Paper folded 4 times is 8mm (equal to 0.008m)

num_layers(21) ➞ "1048.576m"<br>
Paper folded 21 times is 1048576mm (equal to 1048.576m)

**Ans:**

```python
def num_layers(n):
    return (2 ** (n - 1)) / 1000

print(num_layers(1))
print(num_layers(4))
print(num_layers(21))
```

### 4. Create a function that takes a single string as argument and returns an ordered list containing the indices of all capital letters in the string.

**Examples**

index_of_caps("eDaBiT") ➞ [1, 3, 5]<br>
index_of_caps("eQuINoX") ➞ [1, 3, 4, 6]<br>
index_of_caps("determine") ➞ []<br>
index_of_caps("STRIKE") ➞ [0, 1, 2, 3, 4, 5]<br>
index_of_caps("sUn") ➞ [1]

**Ans:**

```python
def index_of_caps(string):
    return [string.index(i) for i in string if i.isupper()]

print(index_of_caps("eDaBiT"))
print(index_of_caps("eQuINoX"))
print(index_of_caps("determine"))
print(index_of_caps("STRIKE"))
print(index_of_caps("sUn"))
```

### 5. Using list comprehensions, create a function that finds all even numbers from 1 to the given number.

**Examples**

find_even_nums(8) ➞ [2, 4, 6, 8]<br>
find_even_nums(4) ➞ [2, 4]<br>
find_even_nums(2) ➞ [2]

**Ans:**

```python
def find_even_nums(n):
    return [i for i in range(1,n+1) if i % 2 == 0]

print(find_even_nums(8))
print(find_even_nums(4))
print(find_even_nums(2))
```
