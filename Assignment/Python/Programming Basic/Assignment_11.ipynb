### 1.	Write a Python program to find words which are greater than given length k?

**Ans:**

```python
k = 4
string ="Python code"

text = string.split(" ")
l = [i for i in text if len(i) > k]        
l
```

### 4.	Write a Python program to check if a given string is binary string or not?

**Ans:**

```python
string = "101010000111"

s = set(string)
print(s)

b = {'0', '1'}

if b == s or s == {'0'} or s == {'1'}:
    print("Yes, it is a binary string")
else :
    print("No, it is not a binary string")
```

### 5.	Write a Python program to find uncommon words from two Strings?

**Ans:**

```python
a = "a b c"
b = "b c d"

a = a.split()
b = b.split()

k = set(a).symmetric_difference(set(b))
print(k)
```

### 6.	Write a Python program to find all duplicate characters in string?

**Ans:**

```python
from collections import Counter
string = 'mississippi'
freq = Counter(string)

print("The duplicate characters in {}: ".format(string))
for k,v in freq.items():
    if v > 1:
        print(k)
```

### 7.	Write a Python Program to check if a string contains any special character?

**Ans:**

```python
import re

string = "Python$Co$de"
regex = re.compile('[@_!#$%^&*()<>?/\|}{~:]')

if(regex.search(string) == None):
    print("Special characters not present")
else:
    print("Special characters present")
```
